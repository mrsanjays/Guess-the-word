#### 2. The `shuffler()` Function



````
# The 'shuffler()' function to to jumble the letters of a word and form a jumbled word.
import random

def shuffler(word):
  letters = list(word)
  random.shuffle(letters)
  shuffled_word = ''.join(letters)
  return(shuffled_word)

shuffler('swim')
````
